```python
import pandas as pd
import random

# Sample values
names = ["Arun", "Priya", "Karthik", "Divya", "Rahul", "Meena", "Vijay", "Anitha", "Suresh", "Kavya"]
genders = ["Male", "Female"]
departments = ["Cardiology", "Neurology", "Orthopedics", "Pediatrics", "General Medicine"]
diseases = ["Diabetes", "Heart Disease", "Fever", "Fracture", "Asthma", "Cancer"]
status = ["Recovered", "Under Treatment", "Discharged"]

# Store records
data = []

# Create 200 records
for i in range(1, 201):
    patient = {
        "Patient_ID": f"P{i:03}",
        "Patient_Name": random.choice(names),
        "Age": random.randint(1, 90),
        "Gender": random.choice(genders),
        "Department": random.choice(departments),
        "Disease": random.choice(diseases),
        "Days_Admitted": random.randint(1, 30),
        "Treatment_Cost": random.randint(5000, 100000),
        "Patient_Status": random.choice(status)
    }

    data.append(patient)

# Convert to DataFrame
df = pd.DataFrame(data)

# Save as CSV file
df.to_csv("hospital_patient_dataset.csv", index=False)

# Display first 10 rows
print("Hospital Patient Dataset Created Successfully!")
print(df.head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("hospital_patient_dataset.csv")

plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=10)
plt.title("Age-wise Patient Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Patients")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("hospital_patient_dataset.csv")

gender_count = df["Gender"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(gender_count, labels=gender_count.index, autopct='%1.1f%%')
plt.title("Gender-wise Patient Distribution")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("hospital_patient_dataset.csv")

department_count = df["Department"].value_counts()

plt.figure(figsize=(8,5))
plt.bar(department_count.index, department_count.values)
plt.title("Department-wise Patient Count")
plt.xlabel("Department")
plt.ylabel("Number of Patients")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("hospital_patient_dataset.csv")

disease_count = df["Disease"].value_counts()

plt.figure(figsize=(8,5))
plt.plot(disease_count.index, disease_count.values, marker='o')
plt.title("Disease-wise Patient Analysis")
plt.xlabel("Disease")
plt.ylabel("Number of Patients")
plt.xticks(rotation=45)
plt.show()
```
